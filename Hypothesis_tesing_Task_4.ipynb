```python
electronics = df[df["Category"]=="Electronics"]["Total_Sales"]

grocery = df[df["Category"]=="Grocery"]["Total_Sales"]
```

```python
print("Electronics:", len(electronics))
print("Grocery:", len(grocery))
```

```python
t_stat, p_value = ttest_ind(electronics, grocery)

print("T-statistic =", t_stat)
print("P-value =", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
    print("There is a statistically significant difference between Electronics and Grocery sales.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("There is NO statistically significant difference between Electronics and Grocery sales.")
```

```python
print("Average Electronics Sales:", electronics.mean())
print("Average Grocery Sales:", grocery.mean())
```

```python
print("Electronics Standard Deviation:", electronics.std())
print("Grocery Standard Deviation:", grocery.std())
```

**Final Summary**

```python
print("========== HYPOTHESIS TESTING SUMMARY ==========")

print("\nNull Hypothesis (H0):")
print("There is no significant difference in average sales between Electronics and Grocery.")

print("\nAlternative Hypothesis (H1):")
print("There is a significant difference in average sales between Electronics and Grocery.")

print("\nAverage Electronics Sales:", electronics.mean())
print("Average Grocery Sales:", grocery.mean())

print("\nT-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("\nConclusion:")
    print("Reject H0.")
    print("Electronics and Grocery sales are significantly different.")
else:
    print("\nConclusion:")
    print("Fail to Reject H0.")
    print("No statistically significant difference was found.")
```
